```python
import pandas as pd
```

```python
train_path = "inference_dataset/inference_dataset/train.tsv"
val_path = "inference_dataset/inference_dataset/dev.tsv"
test_path = "inference_dataset/inference_dataset/test.tsv"


train_df = pd.read_csv(train_path, sep="\t")
test_df = pd.read_csv(test_path, sep="\t")
val_df = pd.read_csv(val_path, sep="\t")
```

```python
import re

def extract_raw_sentence(parsed_text):
    cleaned_text = re.sub(r'\((\S+) ', '', parsed_text)  
    cleaned_text = re.sub(r'\)', '', cleaned_text)  
    return cleaned_text.strip()
```

```python
train_df["Sent1_raw"] = train_df["Sent1_parse"].apply(extract_raw_sentence)
train_df["Sent2_raw"] = train_df["Sent2_parse"].apply(extract_raw_sentence)

train_df[["Sent1_raw", "Sent2_raw", "Label"]].head()
```

```python
val_df["Sent1_raw"] = val_df["Sent1_parse"].apply(extract_raw_sentence)
val_df["Sent2_raw"] = val_df["Sent2_parse"].apply(extract_raw_sentence)

val_df[["Sent1_raw", "Sent2_raw", "Label"]].head()
```

```python
test_df["Sent1_raw"] = test_df["Sent1_parse"].apply(extract_raw_sentence)
test_df["Sent2_raw"] = test_df["Sent2_parse"].apply(extract_raw_sentence)

test_df[["Sent1_raw", "Sent2_raw", "Label"]].head()
```

```python
print(train_df.info())
print(val_df.info())
print(test_df.info())
```

```python
train_df = train_df.drop(["Sent1_parse", "Sent2_parse"], axis = 1)
train_df.head()
```

```python
test_df = test_df.drop(["Sent1_parse", "Sent2_parse"], axis = 1)        
test_df.head()
```

```python
val_df = val_df.drop(["Sent1_parse", "Sent2_parse"], axis = 1)        
val_df.head()
```

```python
train_df.to_csv('train.csv', index=False) 
val_df.to_csv('val.csv', index=False)
test_df.to_csv('test.csv', index=False) 
```

```python
label_mapping = {'entailment': 0, 'contradiction': 1, 'neutral': 2}
train_df['label'] = train_df['Label'].map(label_mapping)
val_df['label'] = val_df['Label'].map(label_mapping)
test_df['label'] = test_df['Label'].map(label_mapping)
```

```python
def get_template(df):
    df = df.dropna(how = "any", subset=["label"])
    
    transformed_text = []
    
    for _, row in df.iterrows():
        premise = row['Sent1_raw'].strip()
        hypothesis = row['Sent2_raw'].strip()
        label = row['label']
        
        formatted = f"Premise: {premise} Hypothesis: {hypothesis}"
        transformed_text.append(formatted)

    transformed_df = pd.DataFrame({'text': transformed_text, 'label': df['label'].tolist()})
    
    return transformed_df
```

```python
val_transformed_df = get_template(val_df)
val_transformed_df.head()
```

```python
train_transformed_df = get_template(train_df)
train_transformed_df.head()
```

```python
test_transformed_df = get_template(test_df)
test_transformed_df.head()
```

```python
print(train_transformed_df.isnull().values.any())
print(test_transformed_df.isnull().values.any())
print(val_transformed_df.isnull().values.any())
```

```python
train_transformed_df.to_csv('train_transformed.csv', index=False)
val_transformed_df.to_csv('val_transformed.csv', index=False)
test_transformed_df.to_csv('test_transformed.csv', index=False)
```
